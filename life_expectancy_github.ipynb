```python
# I see most of problematic entries in 8 variables. so indentiefing typoes outamtically I used 2 std deviation form mean.

for b in lf.Country.unique():
    
    for a in ['Life_expectancy', 'Adult_Mortality', 'BMI', 'Hepatitis_B', 'Polio', 'Diphtheria', 'GDP','infant_deaths']:
        countrystd=lf.loc[lf.Country==b, a].std()
        countryavg=lf.loc[lf.Country==b, a].mean()

        lowerbound=lf.loc[:,a]<countryavg-(countrystd*2)
        upperbound=lf.loc[:,a]<countryavg+(countrystd*2)

        lf.loc[(lf.Country==b) & (lowerbound)&(upperbound), a]= None
```

```python
# ffill loop

for i in lf.columns:
    
    for j in lf.Country.unique():

        if sum(lf.loc[lf.Country == j, i].isna()) != 0:
            
            # fill the NA values with previous year
            lf.loc[lf.Country == j, i] = lf.loc[lf.Country == j, i].fillna(method='ffill')
            
            # if the first year is NA, fill it with next year
            lf.loc[lf.Country == j, i] = lf.loc[lf.Country == j, i].fillna(method='bfill')
            
```

```python
q1=[]
q3=[]
iqr=[]
for i in variable:
    Q1=round((lf[i].quantile(0.25)), 2)
    Q3=round((lf[i].quantile(0.75)), 2)
    IQR=Q3-Q1
    q1.append(Q1)
    q3.append(Q3)
    iqr.append(IQR)

lf_iqr=pd.DataFrame([q1,q3,iqr], columns= variable, index=['Q1','Q3', 'IQR'])

lf_iqr=lf_iqr.T
lf_iqr
```

```python
lower_bound = lf_iqr.Q1-(lf_iqr['IQR'] * 1.5)
upper_bound=lf_iqr.Q3+(lf_iqr['IQR']*1.5)
lf_0= lf[~((lf[variable]<lower_bound)|(lf[variable]>upper_bound)).any(axis=1)]
print(len(lf_0))
print ("The proportion of outliers which exist when compared to the dataframe are: " + str( len(lf_0) / len(lf) ))
```

```python
independent=['infant_deaths', 'Alcohol', 'prtg_expenditure', 'Hepatitis_B',
       'Measles', 'BMI', 'Polio', 'Total_expenditure', 'Diphtheria',
       'HIV/AIDS', 'GDP', 'Population', 'thinness_1/19_years',
       'Income_composition', 'Schooling']
```

```python
personal_habit=['BMI','Alcohol']
country_level=[ 'Income_composition', 'Total_expenditure','GDP','Population', 'Status', 'Schooling']
Immunization =['HIV/AIDS', 'Polio', 'Diphtheria','Hepatitis_B', 'Measles']
```

```python
test_data=lf.loc[lf.Year == 2015]
train_data=lf.loc[lf.Year != 2015]
```

```python
test_log=np.log(test_data[independent]+1)
train_log=np.log(train_data[independent]+1)
```

```python
std_test = {}
std_train = {}

model=LinearRegression()
model.fit(X=train_data[personal_habit], y=train_data["Life_expectancy"])
ypred=model.predict(train_data[personal_habit])
rsqr=r2_score(train_data['Life_expectancy'], ypred)
print('personal_habit-train-r2:', rsqr)

std_train['personal_habit'] = rsqr

ypred=model.predict(test_data[personal_habit])
rsqr=r2_score(test_data['Life_expectancy'], ypred)
print('personal_habit-test-r2:', rsqr)

std_test['personal_habit'] = rsqr

model=LinearRegression()
model.fit(X=train_data[country_level], y=train_data["Life_expectancy"])
ypred=model.predict(train_data[country_level])
rsqr=r2_score(train_data['Life_expectancy'], ypred)
print('country_level-train-r2:', rsqr)

std_train['country_level'] = rsqr

ypred=model.predict(test_data[country_level])
rsqr=r2_score(test_data['Life_expectancy'], ypred)
print('country_level-test-r2:', rsqr)

std_test['country_level'] = rsqr

model=LinearRegression()
model.fit(X=train_data[Immunization], y=train_data["Life_expectancy"])
ypred=model.predict(train_data[Immunization])
rsqr=r2_score(train_data['Life_expectancy'], ypred)
print('Immunization-train-r2:', rsqr)

std_train['immunization'] = rsqr

ypred=model.predict(test_data[Immunization])
rsqr=r2_score(test_data['Life_expectancy'], ypred)
print('Immunization-test-r2:', rsqr)

std_test['immunization'] = rsqr

model=LinearRegression()
model.fit(X=train_data[independent], y=train_data["Life_expectancy"])
ypred=model.predict(train_data[independent])
rsqr=r2_score(train_data['Life_expectancy'], ypred)
print('model with all training feature:', rsqr)

std_train['all_features'] = rsqr

ypred=model.predict(test_data[independent])
rsqr=r2_score(test_data['Life_expectancy'], ypred)
print('model with all testing feature:', rsqr)

std_test['all_features'] = rsqr
```
