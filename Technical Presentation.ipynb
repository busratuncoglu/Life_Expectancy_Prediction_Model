## Step3: Typo Error Fix
I recognized that there are many discrepencies between years for some counties features. The issue is manually detecting typo errors is very hard. to fix it easily, first I used line plot to detect most of problematic entries in 8 variables, then indentiefing typoes autamtically I used 2 std deviation form mean. For example Afganistan Adult_Mortality for 2002 shows unreasonable low value, when we checked in manually I see a typo error.

```python

for b in lf.Country.unique():
    
    for a in ['Life_expectancy', 'Adult_Mortality', 'BMI', 'Hepatitis_B', 'Polio', 'Diphtheria', 'GDP','infant_deaths']:
        countrystd=lf.loc[lf.Country==b, a].std()
        countryavg=lf.loc[lf.Country==b, a].mean()

        lowerbound=lf.loc[:,a]<countryavg-(countrystd*2)
        upperbound=lf.loc[:,a]<countryavg+(countrystd*2)

        lf.loc[(lf.Country==b) & (lowerbound)&(upperbound), a]= None
```

## Step5: Filling Missing

Advenced data cleaning was applied in the model. 
Filling the missing was done on a country basis.
At the beginning forward fill method was used in order not to harm the natural development process of data.
If the first year of faeture empty, this time back fill method was applied.
If the a country feature has no  reading, missing filled with all countries avarage.

For Example, There are missing values at 2008, 2001 and 200 for Hepatitis B in Argentina. With this approach, the Hepatitis B immunization rate increases over the years were able to protect.

```python

lf.loc[lf.Country == 'Argentina', 'Hepatitis_B']
```

```python
# missing fill trial forward_fill
lf.loc[lf.Country == 'Argentina', 'Hepatitis_B'].fillna(method='ffill')
```

```python
# ffill loop

for i in lf.columns:
    
    for j in lf.Country.unique():

        if sum(lf.loc[lf.Country == j, i].isna()) != 0:
            
            # fill the NA values with previous year
            lf.loc[lf.Country == j, i] = lf.loc[lf.Country == j, i].fillna(method='ffill')
            
            # if the first year is NA, fill it with next year
            lf.loc[lf.Country == j, i] = lf.loc[lf.Country == j, i].fillna(method='bfill')
            
```

```python
for i in lf.columns[2:]:
    lf[i] = lf[i].fillna(lf[i].dropna().mean())
```

# Detecting Outliers
Interquartile range was used to detect outliers. According IQR 1260 entries of 2928 appeared as outliears.
The proportion of outliers which exist when compared to the dataframe are 43%. It seems outliers detected with this method as a part of nature of data.

```python
q1=[]
q3=[]
iqr=[]
for i in variable:
    Q1=round((lf[i].quantile(0.25)), 2)
    Q3=round((lf[i].quantile(0.75)), 2)
    IQR=Q3-Q1
    q1.append(Q1)
    q3.append(Q3)
    iqr.append(IQR)

lf_iqr=pd.DataFrame([q1,q3,iqr], columns= variable, index=['Q1','Q3', 'IQR'])

lf_iqr=lf_iqr.T
lf_iqr
```

```python
lower_bound = lf_iqr.Q1-(lf_iqr['IQR'] * 1.5)
upper_bound=lf_iqr.Q3+(lf_iqr['IQR']*1.5)
lf_0= lf[~((lf[variable]<lower_bound)|(lf[variable]>upper_bound)).any(axis=1)]
print(len(lf_0))
print ("The proportion of outliers which exist when compared to the dataframe are: " + str( len(lf_0) / len(lf) ))
```
